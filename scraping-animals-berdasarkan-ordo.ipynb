```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs4

```

```python
ordos_page = requests.get("https://id.wikipedia.org/wiki/Daftar_binatang_menyusui").text
ordos_page
```

```python
ordos_animals_soup = bs4(ordos_page,"lxml")
ordos = ordos_animals_soup.find_all('h3' > 'span', {"class":"mw-headline"})
ordos
```

```python
unwanted_words = ["Daftar","Lihat pula"]
target_ordos = [continent.text for continent in ordos if continent.text not in unwanted_words]
target_ordos
```

```python
ul_html = ordos_animals_soup.find_all('ul')
all_animals = [animals.find_all('li',{"class": None, "id": None}) for animals in ul_html]
all_animals
```

```python
animals_in_ordos = []
for items in all_animals:
    animals = []
    if items:
        for animal in items:
            animals = [animal.find('a').text for animal in items if animal.find('a')]
        animals_in_ordos.append(animals)
animals_in_ordos
```

```python
animals_in_ordos = []
for items in all_animals:
    animals = []
    if items:
        for animal in items:
            animals = [animal.find('a').text for animal in items if animal.find('a')]
        animals_in_ordos.append(animals)
animals_in_ordos
```

```python
animals_ordo_category_df = pd.DataFrame(
    zip(animals_in_ordos, target_ordos), columns=['Animals', 'Ordos'])
animals_ordo_category_df
```

```python
animals_ordo_category_df = animals_ordo_category_df.explode(
    'Animals').reset_index(drop=True)
animals_ordo_category_df
```

```python
animals_ordo_category_df.to_csv('hasil_scraping.csv', index=False)
```
